```python
#!/usr/bin/env python3
# 这个脚本是为了给财务计算总公司和子公司之间的相互借款计算利息用的，只能计算同一年之内的利息，不支持跨年度的结算
# created by weiyq 20200513

import pandas as pd
from datetime import datetime
# pandas禁用科学表达式
pd.set_option('display.float_format',lambda x : '%.3f' % x)
#读取工作簿和工作簿中的工作表
# 运行的时候，需要修改数据的来源，比如天津**资金利息 ，是excel文件的sheet的名字
writer = pd.ExcelWriter('/Users/weiyq/Desktop/lixi/天津**资金利息.xls')
# for sheete_name in sheet_list:
sheete_name='天津**资金利息'
data=pd.read_excel('/Users/weiyq/Desktop/lixi/2019年各公司资金占用3.xls',sheet_name=sheete_name)
data.columns = ['日期','支付金额/元','还款/元','借款金额/元','借款天数/天','借款利率/年','利息金额/元']
data['支付金额/元'].astype('float')
data['还款/元'].astype('float')
data['借款天数/天'].astype('float')
data['借款金额/元'].astype('float')4
data['借款利率/年'].astype('float')
data['利息金额/元'].astype('float')
data = data.fillna(0)
data
# writer.close()
```

```python
# 不是闰年 每个月多少天
days1 = [31,28,31,30,31,30,31,31,30,31,30,31]
# 是闰年 每个月多少天
days2 = [31,29,31,30,31,30,31,31,30,31,30,31]

#计算是不是闰年，
def checkIsRunYear(a):
    if(a % 400 == 0 or  (a % 4 == 0 and a % 100 != 0)):
        return True
    else :
        return False
    
    
#     计算总共借款多少天，算开始的那一天，但是不算截止的那一天
def cal_days(start_time,end_time):
    year = start_time.year
    isRunYear = checkIsRunYear(year)
    start_month = start_time.month
    start_day = start_time.day
    end_month = end_time.month
    end_day = end_time.day
    day_count = 0
    if start_month==end_month:
            return end_day -start_day
        
    if(not isRunYear):
        
        i=start_month
        while(i<=end_month):
            if i==start_month:
                day_count += days1[start_month-1]-start_day+1
                i= i+1
                continue
            if( i==end_month):
                day_count += end_day-1
                i+=1
                continue
            day_count +=  days1[i-1]
            i+=1
    else:
        i=start_month
        while(i<=end_month):
            if i==start_month:
                day_count += days2[start_month-1]-start_day+1
                i+=1
                continue
            if( i==end_month):
                day_count += end_day-1
                i+=1
                continue
            day_count +=  days2[i-1]
            i+=1
    
    return day_count
        
        
        
```

```python


#     从开始借款的那一天开始计算，借款金额，借款天数，借款利息等信息
for index in range(data.shape[0]):

    if index == data.shape[0]-1:
        start = data.loc[index]
        start_time = start['日期']
        s_out = start['支付金额/元']
        s_in = start['还款/元']

#         如果是 本年度的最后一次借款还款的时间
        end_time = datetime(start_time.year,12,31)

    else:
        start = data.loc[index]
        start_time = start['日期']
        s_out = start['支付金额/元']
        s_in = start['还款/元']


        end = data.loc[index+1]
        end_time = end['日期']
    #     e_out = end['支付金额/元']
    #     e_in = end['还款/元']

    lent_money = s_out-s_in
    if index>0:
#         如果不是最后一次借款
        lent_money += data.loc[index-1,'利息金额/元']+data.loc[index-1,'借款金额/元']

    data.loc[index,'借款金额/元'] = lent_money
    lent_days = cal_days(start_time,end_time)
    data.loc[index,'借款天数/天'] = lent_days
    rent_ratio=0
    if lent_money>0:
        #子公司借总公司的钱，不管时间多久年利率6%，
        rent_ratio = 0.06
    else:
#         总公司借子公司的钱，时间少于31天，年利率1.2% ，时间大于等于31天年利率1.4%
        if lent_days <31:
            rent_ratio = 0.012
        else:
    #         
            rent_ratio = 0.014
    data.loc[index,'借款利率/年'] = rent_ratio
    data.loc[index,'利息金额/元'] = lent_money*rent_ratio*lent_days/360
    
data
```

```python
#算一年的利息总和 （总公司需要向子公司支付的利息，小于0 就是子公司需要向总公司支付的利息）

# print("正在保存"+sheete_name)
data = data.append([{}], ignore_index=True)
data = data.fillna(0)
data.loc[data.shape[0],'借款利率/年'] = '总和'
data.loc[data.shape[0],'利息金额/元'] = data['利息金额/元'].sum()

data

```

```python
# 计算结果写入Excel，只能一个sheet写入一张表，不能写入同一个Excel表的不同sheet，数据会覆盖，导致最后只有一个sheet的数据
data.to_excel(writer,sheet_name=sheete_name)
writer.save()

```
